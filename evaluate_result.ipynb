```python
import pandas as pd
```

```python
test_data = pd.read_csv("data/seed_42/data_test.csv")
result_data = pd.read_csv("results/phishing/no_resample/second/result.csv")
```

```python
incorrect_result_idx = {"index": [], "label": [], "pred": [], "text": []}
for idx in range(len(result_data)):
    arr = result_data.iloc[idx]
    if arr[0] != arr[1]:
        incorrect_result_idx["index"].append(idx)
        incorrect_result_idx["label"].append(arr[0])
        incorrect_result_idx["pred"].append(arr[1])
        incorrect_result_idx["text"].append(test_data["TEXT"].iloc[idx])
print("len of incorrect result:", len(incorrect_result_idx["index"]))
```

```python
df = pd.DataFrame(incorrect_result_idx)
```

```python
df.to_csv("compare_result", index=False)
```
